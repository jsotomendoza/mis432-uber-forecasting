```python
# =============================================================================
# MIS 432: AI in Business — Western Washington University
# Module: Uber Demand Forecasting
# Step 1 of 2: Historical Data Simulation & Exploratory Data Analysis
#
# PURPOSE: Before we can build any forecast model, we need historical data to
# learn from. This notebook simulates two years of realistic Friday night Uber
# trip data near WWU's campus, visualizes it, and summarizes key statistics.
# Real Uber data is proprietary, so simulation lets us practice the full
# forecasting workflow with plausible, structured data.
# =============================================================================

# Install/import all required libraries.
# pandas handles our data table (DataFrame), numpy does the math for our
# simulation, and matplotlib draws the chart. All three come pre-installed
# in Google Colab — no pip installs needed.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import date, timedelta

# Fix the random seed so our "simulated" data is reproducible — every time
# you run this notebook you'll get the exact same numbers, which is important
# for a class assignment where your instructor may re-run your code.
np.random.seed(42)


# =============================================================================
# SECTION 1 — SIMULATE TWO YEARS OF FRIDAY NIGHT TRIP DATA
#
# We generate 104 weekly observations (one per Friday, starting 2023-01-06,
# the first Friday of 2023). Each trip count is built from four components
# stacked on top of each other:
#
#   Base trend      — gradual market growth over two years
#   Seasonal bump   — summer is busier; late December is slower
#   School effect   — WWU in session (Sept–May) adds ~15% more trips
#   Random noise    — real data is never perfectly smooth
#
# We also inject a New Year's Eve spike whenever that Friday lands on Dec 31.
# =============================================================================

# --- 1a. Build the date spine (104 consecutive Fridays) ---
# We start on 2023-01-06 (first Friday of 2023) so all 104 dates land on Fridays.
start_date = date(2023, 1, 6)
fridays = [start_date + timedelta(weeks=w) for w in range(104)]

# --- 1b. Linear trend: grow from ~280 trips/Friday to ~380 over 2 years ---
# This represents Uber's expanding market share in Bellingham as more riders
# download the app and more drivers sign up over time.
trend = np.linspace(280, 380, 104)

# --- 1c. Summer seasonality: July & August Fridays run ~20% busier ---
# Bellingham summers draw tourists, outdoor events, and festival crowds,
# all of which spike rideshare demand even as WWU students are gone.
seasonal_bump = np.array([
    0.20 if d.month in (7, 8)          # peak summer
    else -0.25 if d.month == 12 and d.day >= 20 and d.day <= 30  # holiday lull
    else 0.0
    for d in fridays
])

# --- 1d. School-year effect: WWU academic calendar Sept–May adds ~15% ---
# When classes are in session, thousands of students are on campus Friday
# nights going to restaurants, bars, and events — a major demand driver.
# Summer (June–Aug) does NOT get this boost; late May/early June is
# included because finals and move-out week stay busy.
school_bump = np.array([
    0.15 if d.month in (9, 10, 11, 1, 2, 3, 4) or
            (d.month == 12 and d.day <= 19) or
            (d.month == 5) or
            (d.month == 8 and d.day >= 28)   # move-in week
    else 0.0
    for d in fridays
])

# --- 1e. Combine all effects into a single trip count per week ---
# Formula: base trend × (1 + seasonal%) × (1 + school%) + noise
noise = np.random.normal(loc=0, scale=12, size=104)   # ±12 trips of random variation
trips_raw = trend * (1 + seasonal_bump) * (1 + school_bump) + noise

# --- 1f. Inject the New Year's Eve spike ---
# NYE is one of Uber's biggest demand nights nationally. Any Friday that
# falls on Dec 31 gets a significant bump above the normal holiday lull.
nye_multiplier = 2.1   # NYE demand roughly doubles vs. a normal December Friday
for i, d in enumerate(fridays):
    if d.month == 12 and d.day == 31:
        trips_raw[i] = trips_raw[i] * nye_multiplier
        print(f"  ✦ New Year's Eve spike injected on {d} — {int(round(trips_raw[i]))} trips")

# Round to integers (you can't have 0.7 of a trip) and floor at 0
friday_night_trips = np.maximum(0, np.round(trips_raw)).astype(int)

# --- 1g. Assemble the DataFrame ---
# This is the "dataset" we'll hand to our forecast model in the next step.
df = pd.DataFrame({
    "week": pd.to_datetime(fridays),
    "friday_night_trips": friday_night_trips
})

print("\n✅  Data simulation complete — 104 Friday nights generated.\n")


# =============================================================================
# SECTION 2 — VISUALIZE THE DATA
#
# A chart lets us quickly spot trends, seasonal patterns, and outliers that
# a table of numbers would hide. Before fitting ANY model you should always
# plot your data — unexpected patterns here will completely change how you
# approach forecasting.
#
# We overlay:
#   • A vertical dashed line at the start of each summer (June 1)
#   • An annotation arrow pointing to the New Year's Eve spike
# =============================================================================

fig, ax = plt.subplots(figsize=(14, 5))

# --- Main trip-count line ---
ax.plot(df["week"], df["friday_night_trips"],
        color="#1a1a2e", linewidth=1.6, alpha=0.9, zorder=3)

# Subtle shaded area under the line adds visual weight without clutter
ax.fill_between(df["week"], df["friday_night_trips"],
                alpha=0.08, color="#1a1a2e")

# --- Summer start markers ---
# We draw a dashed vertical line every time a new June 1 falls in our window.
# This makes the seasonal pattern immediately visible.
summers = pd.date_range(start="2023-06-01", end="2025-06-01", freq="YS-JUN")
for s in summers:
    if df["week"].min() <= s <= df["week"].max():
        ax.axvline(s, color="#e07b39", linestyle="--", linewidth=1.3,
                   alpha=0.8, zorder=2)
        ax.text(s + pd.Timedelta(days=3), ax.get_ylim()[0] + 5,
                f"Summer\n{s.year}", fontsize=7.5, color="#e07b39", va="bottom")

# --- Annotate the New Year's Eve spike ---
# Find the row where NYE occurs (Dec 31 Friday) and point to it.
nye_rows = df[
    (df["week"].dt.month == 12) & (df["week"].dt.day == 31)
]
for _, row in nye_rows.iterrows():
    ax.annotate(
        f"New Year's Eve\n({row['friday_night_trips']} trips)",
        xy=(row["week"], row["friday_night_trips"]),
        xytext=(row["week"] - pd.Timedelta(weeks=7),
                row["friday_night_trips"] - 55),
        fontsize=8, color="#c0392b",
        arrowprops=dict(arrowstyle="->", color="#c0392b", lw=1.2),
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="#c0392b",
                  alpha=0.85, lw=0.8)
    )

# --- Axis formatting ---
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))
plt.xticks(rotation=30, ha="right", fontsize=8.5)
plt.yticks(fontsize=8.5)

ax.set_title("Friday Night Uber Trips Near WWU — 2 Years of History",
             fontsize=14, fontweight="bold", pad=14)
ax.set_xlabel("Week", fontsize=11, labelpad=8)
ax.set_ylabel("Friday Night Trips", fontsize=11, labelpad=8)

ax.grid(axis="y", linestyle="--", alpha=0.35)
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.savefig("wwu_uber_friday_history.png", dpi=150, bbox_inches="tight")
plt.show()
print("📊  Chart saved as 'wwu_uber_friday_history.png'\n")


# =============================================================================
# SECTION 3 — PRINT FIRST & LAST 10 ROWS + SUMMARY STATISTICS
#
# Inspecting the raw rows confirms the DataFrame is structured correctly
# before we pass it to any model. The summary stats give us a quick sense
# of the range we're forecasting into and identify our historical peak —
# useful for sanity-checking model outputs later ("did my forecast ever
# predict more trips than we've ever actually seen?").
# =============================================================================

print("=" * 52)
print("  FIRST 10 ROWS — earliest Fridays in the dataset")
print("=" * 52)
print(df.head(10).to_string(index=False))

print()
print("=" * 52)
print("  LAST 10 ROWS — most recent Fridays in the dataset")
print("=" * 52)
print(df.tail(10).to_string(index=False))

# --- Summary statistics ---
min_trips  = df["friday_night_trips"].min()
max_trips  = df["friday_night_trips"].max()
avg_trips  = df["friday_night_trips"].mean()
peak_row   = df.loc[df["friday_night_trips"].idxmax()]

print()
print("=" * 52)
print("  SUMMARY STATISTICS (104-week window)")
print("=" * 52)
print(f"  Minimum Friday night trips : {min_trips:>6,}")
print(f"  Maximum Friday night trips : {max_trips:>6,}")
print(f"  Average Friday night trips : {avg_trips:>9,.1f}")
print(f"  Peak week                  : {peak_row['week'].strftime('%Y-%m-%d')}"
      f"  ({max_trips:,} trips)")
print("=" * 52)
print()
print("✅  EDA complete. Next step: build the forecast model.")
```

```python
# =============================================================================
# MIS 432: AI in Business — Western Washington University
# Module: Uber Demand Forecasting
# Step 2 of 2: Feature Engineering, Backtesting & Model Evaluation
#
# PURPOSE: Before trusting any model with a real business prediction, we need
# to know how accurate it is on data it has never seen. This notebook adds
# interpretable features to our dataset, trains a linear regression model on
# 18 months of history, and tests it against the final 6 months — a process
# called backtesting. If the error is acceptable for operational planning,
# we can then use the model to forecast Friday night demand for the sold-out
# show scenario.
#
# ASSUMPTION: The DataFrame `df` already exists in this Colab session from
# Step 1 (MIS432_WWU_Uber_EDA.py). Run that notebook first, or paste the
# simulation code above this section in the same notebook.
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error


# =============================================================================
# STEP A — FEATURE ENGINEERING
#
# Raw dates are not useful inputs for a regression model — the algorithm
# doesn't know what "2023-07-07" means in business terms. We translate
# each date into numeric signals that capture the real-world forces driving
# Friday night Uber demand. Each feature below is a hypothesis about human
# behavior that we're asking the model to test.
# =============================================================================

# --- Feature 1: week_num ---
# A simple integer counter (1 = first week, 104 = last week).
# This captures the long-run TREND: as Uber's Bellingham market matures,
# more drivers and riders join the platform, so trip counts drift upward
# over time regardless of season or school calendar. Without this feature,
# the model would treat every week as equally "old" and miss the growth arc.
df["week_num"] = range(1, len(df) + 1)

# --- Feature 2: is_summer ---
# 1 if the Friday falls in June, July, or August; 0 otherwise.
# Summer in Bellingham means festivals, tourists, outdoor events, and late
# sunsets that keep people out longer — all demand drivers. Conversely, WWU
# students are largely gone, so we need a separate feature (is_school_year)
# to let the model disentangle "summer crowds" from "student crowds."
df["is_summer"] = df["week"].dt.month.isin([6, 7, 8]).astype(int)

# --- Feature 3: is_nye ---
# 1 if the Friday contains December 31st (New Year's Eve); 0 otherwise.
# NYE is a structural outlier — demand spikes far above what any trend or
# seasonal pattern would predict. Without flagging it explicitly, the model
# would either average it into normal December (underestimating NYE) or
# treat it as unexplainable noise. One binary flag is all it takes.
df["is_nye"] = (
    (df["week"].dt.month == 12) & (df["week"].dt.day == 31)
).astype(int)

# --- Feature 4: is_school_year ---
# 1 if the Friday falls during WWU's academic year (Sept–May); 0 otherwise.
# Thousands of students on campus on a Friday night are a massive, predictable
# demand signal: bars, restaurants, house parties, late-night food runs.
# This feature captures that calendar rhythm independently of summer tourism,
# so the model can correctly predict that a rainy October Friday (school=1,
# summer=0) still outperforms a sunny August Friday (school=0, summer=1).
df["is_school_year"] = df["week"].dt.month.isin(
    [9, 10, 11, 1, 2, 3, 4, 5]
).astype(int)
# Edge case: include late August move-in week and December through the 19th
df.loc[
    (df["week"].dt.month == 8) & (df["week"].dt.day >= 28), "is_school_year"
] = 1
df.loc[
    (df["week"].dt.month == 12) & (df["week"].dt.day <= 19), "is_school_year"
] = 1

print("✅  Feature engineering complete. Updated DataFrame columns:")
print(f"    {list(df.columns)}\n")
print(df[["week", "friday_night_trips", "week_num",
          "is_summer", "is_nye", "is_school_year"]].head(8).to_string(index=False))
print()


# =============================================================================
# STEP B — TRAIN / TEST SPLIT
#
# We divide our 104 weeks into two non-overlapping windows:
#
#   Training set  (weeks  1–78):  The 18 months the model LEARNS from.
#   Test set      (weeks 79–104): The 6 months the model has NEVER seen.
#
# This mimics real operational forecasting: you train on everything you know
# up to today, then evaluate on the future. If we trained on all 104 weeks
# and then measured error on the same 104 weeks, the model would look
# artificially accurate — it would just be memorizing the data it already
# saw (overfitting). The held-out test set gives us an honest estimate of
# how the model would perform on a future Friday it hasn't encountered yet,
# which is exactly the Friday-night-show scenario we're building toward.
# =============================================================================

TRAIN_CUTOFF = 78   # last week index (1-based) included in training

train = df[df["week_num"] <= TRAIN_CUTOFF].copy()
test  = df[df["week_num"] >  TRAIN_CUTOFF].copy()

print("=" * 58)
print("  TRAIN / TEST SPLIT")
print("=" * 58)
print(f"  Training set : weeks  1–{TRAIN_CUTOFF}  "
      f"({train['week'].min().strftime('%Y-%m-%d')} → "
      f"{train['week'].max().strftime('%Y-%m-%d')})")
print(f"  Test set     : weeks {TRAIN_CUTOFF+1}–104  "
      f"({test['week'].min().strftime('%Y-%m-%d')} → "
      f"{test['week'].max().strftime('%Y-%m-%d')})")
print(f"  Training rows: {len(train)}   |   Test rows: {len(test)}")
print()
print("  WHY THIS SPLIT?")
print("  The model is trained on everything left of week 78 — 18 months")
print("  of Fridays it can study. Then we ask it to predict the final 6")
print("  months (the test set) as if it were doing live forecasting.")
print("  Because those 26 weeks were never shown during training, the")
print("  error we measure is a realistic estimate of future accuracy —")
print("  not an inflated score from the model 'cheating' on data it")
print("  already memorized.")
print("=" * 58)
print()


# =============================================================================
# STEP C — TRAIN THE MODEL AND EVALUATE ON THE TEST SET
#
# Linear regression fits a straight-line equation of the form:
#
#   trips = β0 + β1·week_num + β2·is_summer + β3·is_nye + β4·is_school_year
#
# Each β coefficient tells us: "holding everything else constant, what is
# the estimated change in Friday night trips for a one-unit increase in
# this feature?" The model learns these coefficients from the training data
# and then applies the same equation to the test set — weeks it has never
# seen — to generate predictions.
# =============================================================================

FEATURES = ["week_num", "is_summer", "is_nye", "is_school_year"]

X_train = train[FEATURES]
y_train = train["friday_night_trips"]

X_test  = test[FEATURES]
y_test  = test["friday_night_trips"]

# --- Train ---
model = LinearRegression()
model.fit(X_train, y_train)

# --- Predict on the held-out test set ---
test = test.copy()
test["predicted_trips"] = model.predict(X_test).round().astype(int)

# --- Print the model's learned coefficients ---
print("=" * 58)
print("  MODEL COEFFICIENTS (what the model learned)")
print("=" * 58)
print(f"  Intercept (base trips at week 0) : {model.intercept_:>8.1f}")
for feat, coef in zip(FEATURES, model.coef_):
    print(f"  {feat:<25}: {coef:>+8.1f} trips")
print()
print("  HOW TO READ THESE:")
print("  • week_num coefficient is the weekly trend slope — each passing")
print("    Friday, Uber demand grows by roughly that many trips.")
print("  • is_summer coefficient shows the summer premium over baseline.")
print("  • is_nye coefficient captures the NYE spike magnitude.")
print("  • is_school_year shows how many extra trips WWU students drive.")
print()

# --- Evaluate: Mean Absolute Error ---
mae = mean_absolute_error(y_test, test["predicted_trips"])

print("=" * 58)
print("  BACKTEST ACCURACY")
print("=" * 58)
print(f"  Mean Absolute Error (MAE): {mae:.1f} trips")
print()
print(f"  ➤  On average, the model's Friday night trip predictions")
print(f"     were off by {mae:.0f} trips.")
print()

# --- Operational interpretation ---
avg_trips = df["friday_night_trips"].mean()
pct_error = (mae / avg_trips) * 100

print("  IS THIS ACCEPTABLE FOR OPERATIONAL PLANNING?")
print(f"  Average Friday night demand is ~{avg_trips:.0f} trips.")
print(f"  An MAE of {mae:.0f} trips = ~{pct_error:.1f}% average error.\n")

if pct_error < 8:
    verdict = (
        "✅  GOOD for operational planning.\n"
        "   Being off by a few percent on a typical Friday means Uber's\n"
        "   dispatch algorithms can compensate in real time. Driver\n"
        "   incentive budgets and surge pricing zones can be set with\n"
        "   confidence at this accuracy level."
    )
elif pct_error < 15:
    verdict = (
        "⚠️   ACCEPTABLE but watch for systematic bias.\n"
        "   At this error level, Uber can still use the forecast for\n"
        "   broad driver supply planning, but should not rely on it\n"
        "   for minute-level dispatch or tight surge pricing decisions.\n"
        "   Adding more features (weather, events) would help."
    )
else:
    verdict = (
        "❌  TOO HIGH for reliable operational use.\n"
        "   An error this large means the model is missing an important\n"
        "   demand driver. Consider adding features like weather, local\n"
        "   events, or using a more flexible model (gradient boosting)."
    )
print(f"  {verdict}")
print()
print("  CONTEXT: Being off by 50 trips on a night with 350 expected is")
print("  very different from being off by 500. A 50-trip error might mean")
print("  a few riders wait 3 extra minutes; a 500-trip error means drivers")
print("  are absent for a crowd of thousands — far more damaging.")
print("=" * 58)
print()


# =============================================================================
# STEP D — VISUALIZE THE BACKTEST
#
# The chart tells the story at a glance: the model sees 18 months of history
# (left of the vertical line), then we compare its predictions (dashed orange)
# against what actually happened (solid cyan) over the final 6 months.
# The ±15% shaded band represents an operationally reasonable tolerance
# window — predictions inside this band are close enough to act on.
# =============================================================================

fig, ax = plt.subplots(figsize=(15, 5.5))

# --- Full actual trip line (all 104 weeks) ---
ax.plot(df["week"], df["friday_night_trips"],
        color="#00bcd4", linewidth=1.8, label="Actual Friday night trips",
        zorder=4)

# --- Predicted line (test set only, weeks 79–104) ---
ax.plot(test["week"], test["predicted_trips"],
        color="#ff8c00", linewidth=2.0, linestyle="--",
        label="Model predictions (test set)", zorder=5)

# --- ±15% tolerance band around predictions ---
# This band shows the range within which Uber's operations team would
# consider the forecast "close enough" to plan driver supply around.
lower = test["predicted_trips"] * 0.85
upper = test["predicted_trips"] * 1.15
ax.fill_between(test["week"], lower, upper,
                alpha=0.18, color="#ff8c00",
                label="±15% tolerance band")

# --- Vertical cutoff line at week 78 ---
cutoff_date = train["week"].max()
ax.axvline(cutoff_date, color="#555555", linewidth=1.4,
           linestyle="-", zorder=3)
ax.text(
    cutoff_date - pd.Timedelta(days=8),
    df["friday_night_trips"].max() * 0.97,
    "Model trained on\neverything left\nof this line",
    ha="right", va="top", fontsize=8, color="#333333",
    bbox=dict(boxstyle="round,pad=0.35", fc="white",
              ec="#aaaaaa", alpha=0.9, lw=0.8)
)

# --- Light shading over the test window so it's immediately obvious ---
ax.axvspan(test["week"].min(), test["week"].max(),
           alpha=0.06, color="#ff8c00", zorder=1)
ax.text(
    test["week"].min() + pd.Timedelta(days=4),
    df["friday_night_trips"].min() * 1.01,
    "← Test window (never seen by model)",
    fontsize=8, color="#cc6600", va="bottom"
)

# --- MAE annotation inside the test window ---
mid_test = test["week"].iloc[len(test)//2]
mid_pred = test["predicted_trips"].iloc[len(test)//2]
ax.annotate(
    f"MAE = {mae:.0f} trips\n({pct_error:.1f}% avg error)",
    xy=(mid_test, mid_pred),
    xytext=(mid_test, mid_pred + 55),
    fontsize=8, color="#cc4400", ha="center",
    arrowprops=dict(arrowstyle="->", color="#cc4400", lw=1.0),
    bbox=dict(boxstyle="round,pad=0.3", fc="white",
              ec="#cc4400", alpha=0.9, lw=0.8)
)

# --- Formatting ---
import matplotlib.dates as mdates
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))
plt.xticks(rotation=30, ha="right", fontsize=8.5)
plt.yticks(fontsize=8.5)

ax.set_title("Friday Night Demand — Backtest Results",
             fontsize=14, fontweight="bold", pad=14)
ax.set_xlabel("Week", fontsize=11, labelpad=8)
ax.set_ylabel("Friday Night Trips", fontsize=11, labelpad=8)

ax.legend(loc="upper left", fontsize=8.5, framealpha=0.9)
ax.grid(axis="y", linestyle="--", alpha=0.3)
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.savefig("wwu_uber_backtest.png", dpi=150, bbox_inches="tight")
plt.show()
print("📊  Backtest chart saved as 'wwu_uber_backtest.png'")
print()
print("=" * 58)
print("  NEXT STEP")
print("=" * 58)
print("  The model is validated. To forecast the sold-out show Friday,")
print("  create a one-row DataFrame with the event date's features")
print("  (week_num, is_summer, is_nye, is_school_year) and call")
print("  model.predict() — the backtest error above is your ±margin.")
print("=" * 58)
```

```python
# =============================================================================
# MIS 432: AI in Business — Western Washington University
# Module: Uber Demand Forecasting
# Step 3 of 3: Concert Friday Prediction & Operations Decision
#
# PURPOSE: The model is validated (Step 2). Now we use it to answer a real
# business question: how should Uber staff and price for a sold-out Death Cab
# for Cutie / Sleater-Kinney / ODESZA homecoming concert at WWU this Friday?
# This is the moment forecasting becomes operations — numbers become decisions.
#
# ASSUMPTION: The trained `model` object and `df` DataFrame from Steps 1 & 2
# are already in memory in this Colab session. Run those notebooks first.
# =============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Operational constants — adjust these if the manager's assumptions change
TRIPS_PER_DRIVER_PER_HOUR = 3   # each driver completes ~3 rides/hr at peak
BASELINE_DRIVERS_ONLINE   = 25  # typical Friday night Uber supply near WWU
SURGE_THRESHOLD_DRIVERS   = 25  # if we need MORE than this, activate surge

# Feature names must match what the model was trained on in Step 2
FEATURES = ["week_num", "is_summer", "is_nye", "is_school_year"]


# =============================================================================
# STEP A — BUILD THE CONCERT FRIDAY FEATURE ROW
#
# A linear regression model only understands numbers, not dates or event
# names. We translate everything we know about this specific Friday into
# the four numeric features the model was trained on. Think of this row as
# the model's "briefing document" — everything it needs to make a prediction.
# =============================================================================

concert_friday = pd.DataFrame([{
    "week_num"       : 105,   # one week beyond our 104-week training window
    "is_summer"      : 0,     # school is in session — not a summer Friday
    "is_nye"         : 0,     # not New Year's Eve
    "is_school_year" : 1,     # WWU academic calendar: Sept–May = 1
}])

print("=" * 62)
print("  STEP A — CONCERT FRIDAY FEATURE VECTOR")
print("=" * 62)
print()
print("  Feature          Value   Meaning")
print("  " + "-" * 58)
print("  week_num           105   Week 105 is one step beyond our dataset.")
print("                          The model extrapolates the growth trend")
print("                          it learned from weeks 1–104 forward by")
print("                          one additional week.")
print()
print("  is_summer            0   The concert falls during the school year,")
print("                          not in June/July/August. No summer tourist")
print("                          premium — but the school-year effect below")
print("                          more than compensates.")
print()
print("  is_nye               0   This is not New Year's Eve. The NYE spike")
print("                          coefficient learned in training will NOT")
print("                          apply here. Any concert lift is outside")
print("                          the model — we'll flag that separately.")
print()
print("  is_school_year       1   WWU is in full session. Thousands of")
print("                          students on campus on a Friday night is")
print("                          the single biggest demand multiplier in")
print("                          our model. This is already 'on'.")
print()
print("  ⚠️  NOTE: The model does NOT have a feature for 'sold-out concert'.")
print("  The prediction below is a baseline school-year Friday. The concert")
print("  crowd is an unmodeled upside — treat the upper bound as conservative.")
print()


# =============================================================================
# STEP B — GENERATE THE PREDICTION AND CONFIDENCE RANGE
#
# The model returns a point estimate — its single best guess. But we know
# from our Step 2 backtest that predictions carry error (the MAE). We use
# ±15% bounds as our operational planning range: the lower bound helps us
# avoid over-staffing; the upper bound helps us avoid being caught short.
# For a sold-out concert, the operations manager should plan toward the upper
# bound — the downside of too few drivers (stranded fans, bad press) is
# worse than the downside of too many (a few underutilized drivers).
# =============================================================================

predicted_raw   = model.predict(concert_friday[FEATURES])[0]
predicted_trips = round(predicted_raw)
lower_bound     = round(predicted_trips * 0.85)
upper_bound     = round(predicted_trips * 1.15)

print("=" * 62)
print("  STEP B — DEMAND PREDICTION")
print("=" * 62)
print()
print(f"  Predicted Friday night trips : {predicted_trips:>5,}")
print(f"  Lower bound (−15%)           : {lower_bound:>5,}")
print(f"  Upper bound (+15%)           : {upper_bound:>5,}")
print()
print(f"  ➤  Predicted Friday night trips: {predicted_trips:,}")
print(f"     (range: {lower_bound:,} to {upper_bound:,})")
print()
print("  The ±15% range reflects our Step 2 MAE-validated tolerance band.")
print("  For a normal Friday we'd plan to the midpoint. For a sold-out")
print("  concert, we recommend planning toward the upper bound.")
print()


# =============================================================================
# STEP C — DRIVER RECOMMENDATION
#
# We convert trip demand into driver headcount using a simple operational
# ratio: each driver completes approximately 3 trips per hour during a
# post-concert rush (short rides, high density, low repositioning time).
# We calculate three staffing levels so the operations manager can see
# the full range and make an informed judgment call.
# =============================================================================

# Drivers needed = trips ÷ trips_per_driver_per_hour
# We use ceiling division (round up) because you can't have half a driver —
# and it's always better to have one extra than to be one short at 10:01pm.
import math

drivers_predicted = math.ceil(predicted_trips / TRIPS_PER_DRIVER_PER_HOUR)
drivers_lower     = math.ceil(lower_bound     / TRIPS_PER_DRIVER_PER_HOUR)
drivers_upper     = math.ceil(upper_bound     / TRIPS_PER_DRIVER_PER_HOUR)

print("=" * 62)
print("  STEP C — DRIVER STAFFING RECOMMENDATION")
print("=" * 62)
print()
print(f"  Assumption: {TRIPS_PER_DRIVER_PER_HOUR} trips/driver/hour during the post-show rush")
print(f"  Calculation: drivers needed = trips ÷ {TRIPS_PER_DRIVER_PER_HOUR} (rounded up)")
print()
print(f"  Scenario            Trips    Drivers Needed")
print(f"  " + "-" * 44)
print(f"  Conservative (−15%) {lower_bound:>6,}    {drivers_lower:>5,} drivers")
print(f"  Base forecast       {predicted_trips:>6,}    {drivers_predicted:>5,} drivers")
print(f"  Upside   (+15%)     {upper_bound:>6,}    {drivers_upper:>5,} drivers")
print()
print(f"  ➤  Recommendation: target {drivers_predicted}–{drivers_upper} drivers online near")
print(f"     campus by 9:45pm to handle the 10:00pm crowd surge.")
print()
print(f"  Typical baseline supply on a Friday night: {BASELINE_DRIVERS_ONLINE} drivers.")
print(f"  Additional drivers needed above baseline : {max(0, drivers_upper - BASELINE_DRIVERS_ONLINE)}")
print()


# =============================================================================
# STEP D — SURGE PRICING DECISION
#
# Surge pricing serves two functions at once:
#   1. DEMAND SIGNAL — higher prices discourage non-urgent rides,
#      reserving capacity for riders who really need it right now.
#   2. SUPPLY SIGNAL — higher fares attract more drivers to log on
#      and reposition toward campus, increasing supply in real time.
#
# Uber's rule of thumb: if predicted demand requires more drivers than
# are typically online, activate surge BEFORE the event ends so drivers
# have time to arrive. A surprise surge at 10:01pm is too late.
# =============================================================================

surge_needed = drivers_predicted > SURGE_THRESHOLD_DRIVERS

print("=" * 62)
print("  STEP D — SURGE PRICING DECISION")
print("=" * 62)
print()
print(f"  Drivers needed (base forecast): {drivers_predicted}")
print(f"  Typical drivers online tonight: {BASELINE_DRIVERS_ONLINE}")
print(f"  Surge threshold               : >{SURGE_THRESHOLD_DRIVERS} drivers required")
print()

if surge_needed:
    gap = drivers_predicted - BASELINE_DRIVERS_ONLINE
    print(f"  SURGE RECOMMENDATION: ✅  YES — ACTIVATE SURGE PRICING")
    print()
    print(f"  Reason: Base forecast requires {drivers_predicted} drivers but only")
    print(f"  {BASELINE_DRIVERS_ONLINE} are typically online — a gap of {gap} drivers.")
    print(f"  Surge pricing should be pre-activated by 9:30pm so drivers")
    print(f"  can reposition to campus before the 10:00pm crowd exits.")
    surge_word = "YES"
    surge_reason = (
        f"Predicted demand ({predicted_trips} trips) requires {drivers_predicted} drivers, "
        f"exceeding the typical Friday supply of {BASELINE_DRIVERS_ONLINE} — "
        f"surge pricing is needed to attract the additional {gap} drivers and moderate demand."
    )
else:
    print(f"  SURGE RECOMMENDATION: ❌  NO — DO NOT ACTIVATE SURGE")
    print()
    print(f"  Reason: Base forecast requires {drivers_predicted} drivers. The typical")
    print(f"  Friday supply of {BASELINE_DRIVERS_ONLINE} drivers is sufficient to cover")
    print(f"  predicted demand without price escalation. Monitor in real time.")
    surge_word = "NO"
    surplus = BASELINE_DRIVERS_ONLINE - drivers_predicted
    surge_reason = (
        f"Predicted demand ({predicted_trips} trips) requires only {drivers_predicted} drivers, "
        f"within the typical Friday supply of {BASELINE_DRIVERS_ONLINE} — "
        f"no surge needed, but monitor the 10pm exit closely."
    )
print()


# =============================================================================
# STEP E — OPERATIONS MEMO
#
# The forecast and recommendation are only useful if they reach the decision-
# maker in a clear, actionable format. We print a short memo the analyst can
# copy directly into an email or Slack message to the operations manager.
# Good data science includes the last mile: communicating results in the
# language of the business, not the language of the model.
# =============================================================================

from datetime import date, timedelta

# The concert is the next Friday after our dataset ends
concert_date = (pd.to_datetime(df["week"].max()) + pd.Timedelta(weeks=1)).strftime("%A, %B %d, %Y")

memo = f"""
{'=' * 62}
  OPERATIONS MEMO — UBER BELLINGHAM / WWU AREA
{'=' * 62}

  TO   : Operations Manager
  FROM : Demand Forecasting Analyst
  RE   : Driver Supply & Surge Recommendation — Concert Friday
  DATE : {date.today().strftime("%B %d, %Y")} (Thursday morning)

  EVENT
  ─────
  Death Cab for Cutie / Sleater-Kinney / ODESZA homecoming concert
  at WWU. Sold-out show. Expected crowd exits at 10:00 PM Friday.

  DEMAND FORECAST
  ───────────────
  Predicted Friday night trips (campus area): {predicted_trips:,}
  Planning range                            : {lower_bound:,} – {upper_bound:,} trips
  Model basis: linear regression trained on 18 months of Friday
  night data; backtested MAE within acceptable operational range.

  DRIVER STAFFING RECOMMENDATION
  ───────────────────────────────
  Conservative (low demand) : {drivers_lower:>3} drivers online by 9:45 PM
  Base forecast             : {drivers_predicted:>3} drivers online by 9:45 PM
  Upside (high demand)      : {drivers_upper:>3} drivers online by 9:45 PM

  Recommend targeting {drivers_predicted}–{drivers_upper} drivers staged near campus
  before the show ends. Typical Friday baseline is {BASELINE_DRIVERS_ONLINE} drivers.

  SURGE PRICING: {surge_word}
  {'─' * (16 + len(surge_word))}
  {surge_reason}

  UPSIDE RISK
  ───────────
  The model was trained on ordinary school-year Fridays and does
  NOT account for sold-out concert effects. If the show draws
  fans from outside Bellingham, or if the event runs long, demand
  could push 20–30% above the upper bound — consider a contingency
  driver incentive bonus active from 9:45–11:30 PM.

  ACTION ITEMS
  ────────────
  □ Send driver push notification by 6:00 PM Friday (4 hrs notice)
  □ Pre-position surge multiplier in pricing system by 9:30 PM
  □ Monitor live trip requests at 10:00 PM; escalate if queue > 15
  □ Debrief actual vs. predicted trips Monday for model recalibration

{'=' * 62}
"""

print(memo)


# =============================================================================
# VISUALIZATION — Forecast in context of full history
#
# Plot the 104 weeks of history plus the concert Friday prediction as a
# single standout dot, so the operations manager can see at a glance where
# tonight's forecast sits relative to historical demand.
# =============================================================================

fig, ax = plt.subplots(figsize=(15, 5.5))

# Historical actuals
ax.plot(df["week"], df["friday_night_trips"],
        color="#00bcd4", linewidth=1.6, alpha=0.85,
        label="Historical Friday night trips", zorder=3)
ax.fill_between(df["week"], df["friday_night_trips"],
                alpha=0.07, color="#00bcd4")

# Concert Friday — point estimate with error bar
concert_date_dt = pd.to_datetime(df["week"].max()) + pd.Timedelta(weeks=1)
ax.errorbar(concert_date_dt, predicted_trips,
            yerr=[[predicted_trips - lower_bound], [upper_bound - predicted_trips]],
            fmt="o", color="#ff3d3d", markersize=10, capsize=6, capthick=2,
            linewidth=2, label=f"Concert Friday prediction ({predicted_trips} trips)",
            zorder=6)

# Shaded uncertainty band
ax.fill_between([concert_date_dt - pd.Timedelta(days=2),
                 concert_date_dt + pd.Timedelta(days=2)],
                lower_bound, upper_bound,
                alpha=0.25, color="#ff3d3d")

# Surge threshold line
drivers_to_trips_threshold = SURGE_THRESHOLD_DRIVERS * TRIPS_PER_DRIVER_PER_HOUR
ax.axhline(drivers_to_trips_threshold, color="#ff8c00", linewidth=1.2,
           linestyle=":", alpha=0.7,
           label=f"Surge threshold ({drivers_to_trips_threshold} trips = {SURGE_THRESHOLD_DRIVERS} drivers)")

# Annotation
ax.annotate(
    f"Concert Friday\n{predicted_trips} trips predicted\nSurge: {surge_word}",
    xy=(concert_date_dt, predicted_trips),
    xytext=(concert_date_dt - pd.Timedelta(weeks=8), predicted_trips + 35),
    fontsize=8.5, color="#cc0000", ha="center",
    arrowprops=dict(arrowstyle="->", color="#cc0000", lw=1.2),
    bbox=dict(boxstyle="round,pad=0.4", fc="white", ec="#cc0000",
              alpha=0.92, lw=1.0)
)

import matplotlib.dates as mdates
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b '%y"))
plt.xticks(rotation=30, ha="right", fontsize=8.5)
plt.yticks(fontsize=8.5)

ax.set_title("Friday Night Uber Demand — Concert Forecast vs. 2-Year History",
             fontsize=14, fontweight="bold", pad=14)
ax.set_xlabel("Week", fontsize=11, labelpad=8)
ax.set_ylabel("Friday Night Trips", fontsize=11, labelpad=8)

ax.legend(loc="upper left", fontsize=8.5, framealpha=0.92)
ax.grid(axis="y", linestyle="--", alpha=0.3)
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.savefig("wwu_uber_concert_forecast.png", dpi=150, bbox_inches="tight")
plt.show()
print("📊  Forecast chart saved as 'wwu_uber_concert_forecast.png'")
print()
print("✅  All three steps complete.")
print("    Step 1: Simulated & explored 2 years of Friday night data")
print("    Step 2: Feature-engineered, trained, and backtested the model")
print("    Step 3: Predicted concert Friday demand & issued ops recommendation")
```
